```python
from sklearn.svm import SVC

svc_clf = SVC(random_state=2)
svc_clf.fit(under_x_train, under_y_train)

svc_clf_pred_under = svc_clf.predict(under_x_test)
print('test svc precision: ',precision_score(under_y_test, svc_clf_pred_under))
print('test svc Recall: ',recall_score(under_y_test, svc_clf_pred_under))

svc_clf_pred_under = svc_clf.predict(clone.iloc[:,:-1])
print('data svc precision: ',precision_score(clone.iloc[:,-1], svc_clf_pred_under))
print('data svc Recall: ',recall_score(clone.iloc[:,-1], svc_clf_pred_under))
```

```python
from sklearn.linear_model import LogisticRegression

log_clf = LogisticRegression(random_state=2)
log_clf.fit(under_x_train, under_y_train)

log_clf_pred_under = log_clf.predict(under_x_test)
print('test log precision: ',precision_score(under_y_test, log_clf_pred_under))
print('test log Recall: ',recall_score(under_y_test, log_clf_pred_under))

log_clf_pred_under = log_clf.predict(clone.iloc[:,:-1])
print('test log precision: ',precision_score(clone.iloc[:,-1], log_clf_pred_under))
print('test log Recall: ',recall_score(clone.iloc[:,-1], log_clf_pred_under))
```
